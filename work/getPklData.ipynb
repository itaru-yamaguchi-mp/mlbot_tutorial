```python
import pandas as pd
from crypto_data_fetcher.gmo import GmoFetcher
```

```python
# GMOコインAPIを使う
fetcher = GmoFetcher()
```

```python
import logging
logging.basicConfig(level=logging.DEBUG)

print("Fetching data...")
try:
    df = fetcher.fetch_ohlcv(
        market='BTC_JPY',
        interval_sec=15 * 60,
    )
    print("Data fetched successfully!")
except Exception as e:
    print(f"Error: {e}")
```

```python
# 実験に使うデータ期間を限定する（2024年12月1日～2025年2月4日）
df = df[(pd.to_datetime('2024-12-01 00:00:00Z') < df.index) & (df.index < pd.to_datetime('2025-02-04 00:00:00Z'))]
df.to_pickle('df_ohlcv.pkl')
```
